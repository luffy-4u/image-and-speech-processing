```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
img=plt.imread('vanacum.webp')
print("original image",img.shape)
plt.subplot(1,2,1)
plt.imshow(img)
plt.title('original image')
#
pixel=img.reshape(-1,3)
pixel=np.float32(pixel)
#applying kmeans clustering
criteria=(cv2.TERM_CRITERIA_EPS+cv2.TERM_CRITERIA_MAX_ITER,100,0.85)
k=3
#RETURN
_,label,centre=cv2.kmeans(pixel,k ,criteria,None,10,cv2.KMEANS_RANDOM_CENTERS)
segmented_pixels=(centre[label.flatten()])
segmented_image=segmented_pixels.reshape(img.shape)
segmented_image = segmented_image / 255.0
#plot the final op
plt.subplot(1,2,2)
plt.title('segmented image')
plt.imshow(segmented_image)
plt.show()
```
