```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img=cv2.imread('vanacum.webp')
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
R,G,B=cv2.split(img_rgb)
img_ycrcb=cv2.cvtColor(img_rgb,cv2.COLOR_RGB2YCrCb)
Y,Cr,Cb=cv2.split(img_ycrcb)
img_hsv=cv2.cvtColor(img_rgb,cv2.COLOR_RGB2HSV)
H,S,I=cv2.split(img_hsv)
fig,axs= plt.subplots(3,4,figsize=(12,6))
axs[0,0].imshow(R,cmap='gray')
axs[0,0].set_title("red")
axs[0,0].axis('off')
axs[0,1].imshow(G,cmap='gray')
axs[0,1].set_title("Green")
axs[0,1].axis('off')
axs[0,2].imshow(B,cmap='gray')
axs[0,2].set_title("blue")
axs[0,2].axis('off')
axs[0,3].imshow(Y,cmap='gray')
axs[0,3].set_title("luma")
axs[0,3].axis('off')
axs[1,0].imshow(Cr,cmap='gray')
axs[1,0].set_title("chromite red")
axs[1,0].axis('off')
axs[1,1].imshow(Cb,cmap='gray')
axs[1,1].set_title("chromite blue")
axs[1,1].axis('off')
axs[1,2].imshow(H,cmap='gray')
axs[1,2].set_title("hue")
axs[1,2].axis('off')
axs[1,3].imshow(S,cmap='gray')
axs[1,3].set_title("Saturation")
axs[1,3].axis('off')
axs[2,0].imshow(I,cmap='gray')
axs[2,0].set_title("intensity")
axs[2,0].axis('off')
plt.tight_layout()
plt.show()
```
